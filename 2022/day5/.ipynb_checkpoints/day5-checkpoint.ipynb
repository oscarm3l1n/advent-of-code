```python
if debug:
    with open('test.in', 'r') as f:
        data = f.read().split('\n')
else:
    with open('input.in', 'r') as f:
        data = f.read().split('\n')
```

```python
data = [(int(i.split()[1]), int(i.split()[3]), int(i.split()[5])) for i in data]
```

```python
data
```

```python
if debug:
    stacks = [
        ['z', 'n'],
        ['m','c','d'],
        ['p']
    ]
else:
    stacks = [
        ['q','m','g','c','l'],
        ['r','d','l','c','t','f','h','g'],
        ['v','j','f','n','m','t','w','r'],
        ['j','f','d','v','q','p'],
        ['n','f','m','s','l','b','t'],
        ['r','n','v','h','c','d','p'],
        ['h','c','t'],
        ['g','s','j','v','z','n','h','p'],
        ['z','f','h','g']
    ]
```

```python
for n_box, from_, to in data:
    for i in range(n_box):
        box = stacks[from_-1].pop()
        stacks[to-1].append(box)

res = [stack.pop() for stack in stacks]

print("".join(res).upper())
```

```python
if debug:
    stacks = [
        ['z', 'n'],
        ['m','c','d'],
        ['p']
    ]
else:
    stacks = [
        ['q','m','g','c','l'],
        ['r','d','l','c','t','f','h','g'],
        ['v','j','f','n','m','t','w','r'],
        ['j','f','d','v','q','p'],
        ['n','f','m','s','l','b','t'],
        ['r','n','v','h','c','d','p'],
        ['h','c','t'],
        ['g','s','j','v','z','n','h','p'],
        ['z','f','h','g']
    ]
```

```python
for n_box, from_, to in data:
    curr_stack = stacks[from_ - 1]
    boxes = curr_stack[len(curr_stack) - n_box:]
    # remove boxes from arr
    for i in range(n_box):
        stacks[from_ - 1].pop()
    for box in boxes:
        stacks[to - 1].append(box)
        
res2 = [stack.pop() for stack in stacks]


print("".join(res2).upper())
```
